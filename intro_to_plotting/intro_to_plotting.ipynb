**Problem 1**

```python
a = 1
b = a
a *= 2
print(a)
print(b)
```

The \*= operator multiplies the right value by the left value and assigns the product to the left operand. So, a \*= 2 is equivalent to a = a * 2.

**Problem 2**

```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 3, 5, 5])
b = a
a = a * 2
print(a)
print(b)
```

In the previous problem, b was assigned to a. When a was modified by the \*= operator, b stayed the same. However, in this problem, when a was modified by the operator, b was modified as well. 

This happened because the \*= operator invokes the \__imul\__ function which changes the value of the left operand in-place, whereas the * operator invokes the \__mul\__ function which changes the location of the left operand in memory. When we print a variable like b, which was assigned before either function was invoked, we see the difference between the two. 

When \*= was used, the original value of a was printed because the variable assignment still refers to the original location. When * was used, the new value of a was printed because the variable was reassigned to a different location.

**Problem 4**

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plotting y = 3 cos(x)

x = np.linspace(0, 360, 500)
y = 3 * np.cos(np.deg2rad(x))
plt.xlabel('x (degrees)')
plt.ylabel('3 cos(x)')
plt.title('Graph of f(x)')
plt.plot(x, y, 'b-')
plt.show()
```

```python
# plotting y = 2 cos(2x)

x = np.linspace(0, 360, 500)
y = 2 * np.cos(2 * (np.deg2rad(x)))
plt.xlabel('x (degrees)')
plt.ylabel('2 cos(2x)')
plt.title('Graph of f(x)')
plt.plot(x, y, 'b--')
plt.show()
```

```python
# plotting y = cos(3x)

x = np.linspace(0, 360, 500)
y = np.cos(3 * (np.deg2rad(x)))
plt.xlabel('x (degrees)')
plt.ylabel('cos(3x)')
plt.title('Graph of f(x)')
plt.plot(x, y, 'b+')
plt.show()
```

**Problem 5**

```python
# plotting f(x) = x and g(x) = rx(1 - x)
# for varying values of r

x = np.linspace(0, 1, 100)
r = [0.5, 1.5, 2.5, 3.5, 4]
f = x
g = [(r_val * x * (1 - x)) for r_val in r]
for func in range(0, len(g)):
    plt.plot(f, g[func])
    plt.legend([0.5, 1.5, 2.5, 3.5, 4])
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x) = x, g(x) = rx(1-x)')
plt.title('Graphs of f(x) and g(x)')
plt.show()
```

The curves of f(x) and g(x) intersect for all values of r at x = 0 and x = 1. At r = 1.5, the curves also intersect at x = 1/3.

**Problem 6**

```python
# plotting g(x) = rx(1 - x) 
# and h(x) = r g(x) (1 - g(x))
# for varying values of r

x = np.linspace(0, 1, 100)
r = [0.5, 1.5, 2.5, 3.5, 4]
h = [(r_val * (r_val * x * (1 - x)) * (1 - (r_val * x * (1 - x)))) for r_val in r]
for func in range(0, len(h)):
    plt.plot(x, h[func])
    plt.legend([0.5, 1.5, 2.5, 3.5, 4])
plt.xlabel('x')
plt.ylabel('h(g(x))')
plt.title('Graphs of h(x)')
plt.show()
```

**Problem 7**

```python
# plotting the world's CO2 record

data = np.loadtxt('ice_core_CO2_data.txt', usecols = (0, 1))

plt.rcParams['figure.figsize'] = [15, 5]
plt.plot(data[:,0], data[:,1], 'r-')

CO2 = data[:,1]
average = sum(CO2) / CO2.shape[0]
plt.fill_between(data[:,0], data[:,1], average, color=(1.0,0.90,0.90), lw=0)

plt.xlabel('Years before 1950')
plt.ylabel('CO2 (ppmv)')
plt.title('World\'s CO2 data')
plt.show()
```
