```python
n = 11
data = np.zeros((n, n))

for i in range(0, n):
    data[i, i] = 1
    data[0:i, i] = 1
    
plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n, n))

for i in range(0, n//2):
    data[i, n // 2 - i] = 1
    data[n // 2 - i, i] = 1
    data[n // 2 + i, i] = 1
    data[i, n // 2 + i] = 1
    data[n - 1 - i, n // 2 + i] = 1
    data[n // 2, n - 1] = 1

plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 51
data = np.zeros((n, n))

for i in range(0, n):
    data[i, i] = 1
    data[i, (n - 1) - i] = 1
    
for j in range(0, n // 2):
    data[j, j: n - j] = 1

for k in range(n // 2, n):
    data[k, n- k:k+1] = 1

plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n, n))

for i in range (0, n):
    for j in range(0, n):
        if i % 2 == 0:
            if j % 2 == 0:
                data[i, j] = 1
        if i % 2 == 1:
            if j % 2 == 1:
                data[i, j] = 1
                
plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 13
data = np.zeros((n, n))

for j in range(0, n + 1, 3):  
    for i in range(0, j + 1):
        data[i, j - i] = 1

for j in range(3, n + 1, 3):  
    for i in range(12, j - 1, -1):
        data[i, j - 1 - i] = 1
        
plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 501
r = n // 2
data = np.zeros((n, n))
 
for i in range(0, r):
    for j in range(0, r):
        if np.linalg.norm([i, j]) <= r:
            data[i, j] = 1

for i in range(0, r):
    for j in range(r, n):
        if np.linalg.norm([i, j - n - 1]) <= r:
            data[i, j] = 1

for i in range(r, n):
    for j in range(0, r):
        if np.linalg.norm([i - n - 1, j]) <= r:
            data[i, j] = 1
            
for i in range(r, n):
    for j in range(r, n):
        if np.linalg.norm([i - n  - 1, j - n - 1]) <= r:
            data[i, j] = 1

plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 501
r = n // 2
data = np.zeros((n, n))

top = 100
bottom = 400

center1 = 150
center2 = 350
 
for i in range(top, bottom):
    for j in range(0, ((n // 3) * 2)):
        if np.linalg.norm([i - r, j - (n // 3)]) <= center1:
            data[i, j] = 1
    for j in range ((n // 3), n):   
        if np.linalg.norm([i - r, j - center2]) <= center1:
            data[i, j] = 1
    for j in range(0, n):
        if np.linalg.norm([i - r, j - (n // 3)]) <= center1 and np.linalg.norm([i - r, j - center2]) <= center1:
            data[i, j] = 0
plt.imshow(data, interpolation = 'nearest')
plt.show()
```
