```python
n = 11
data = np.zeros((n, n))

for i in range(0, n):
    data[i, i] = 1
    data[i, (n - 1) - i] = 1

plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n, n))

for i in range(1, n - 1):
    data[1, i] = 1
    data[n - 2, i] = 1
    data[i, 1] = 1
    data[i, n - 2] = 1
    
plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n, n))

for i in range(0, n):
    data[0, i] = 1
    data[i, 0] = 1
    data[n - 1, i] = 1
    data[i, n - 1] = 1
    
for i in range(2, n - 2):
    data[i, 2] = 1
    data[2, i] = 1
    data[i, n - 3] = 1
    data[n - 3, i] = 1 

for i in range(4, n - 4):
    data[i, 4] = 1
    data[4, i] = 1
    data[i, n - 5] = 1
    data[n - 5, i] = 1 
plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n, n))

for i in range (1, 5):
    for j in range(1, 5):
        data[i, j] = 1

for i in range (6, 10):
    for j in range(6, 10):
        data[i, j] = 1
        
plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n, n))

for i in range (1, (n - 1) // 2):
    for j in range(1, (n - 1) // 2):
        data[i, j] = 1

for i in range (1, (n - 1) // 2):
    for j in range(((n - 1) // 2) + 1, n - 1):
        data[i, j] = 1
        
for i in range (((n - 1) // 2) + 1, n - 1):
    for j in range(1, (n - 1) // 2):
        data[i, j] = 1

for i in range (((n - 1) // 2) + 1, n - 1):
    for j in range(((n - 1) // 2) + 1, n - 1):
        data[i, j] = 1

plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n, n))

data[(n - 1) // 2, 1] = 1

for i in range(4, n - 4):
    data[i, 3] = 1

for i in range(3, n - 3):
    data[i, 5] = 1

for i in range(2, n - 2):
    data[i, 7] = 1
    
for i in range(1, n - 1):
    data[i, 9] = 1
    
plt.imshow(data, interpolation = 'nearest')
plt.show()
```

```python
n = 501
r = n / 2.0
data = np.zeros((n, n))
 
plt.imshow(data, interpolation = 'nearest')
for i in range(0, n):
    for j in range(0, n):
        if np.linalg.norm([i - ((n - 1) // 2), j - ((n - 1) // 2)]) <= r:
            data[i, j] = 1
            
plt.imshow(data, interpolation = 'nearest')
plt.show()
```

**Problem 3**

```python
# Earth's elevation data
h = np.loadtxt('topography_180x360_grid.txt')
h.shape
plt.imshow(h, interpolation = 'nearest')
plt.show()
```

```python
# covering Australia
h2 = h
for i in range(75, 150):
    for j in range(125, 175):
        h2[i, j] = 10000
plt.imshow(h2, interpolation = 'nearest')
plt.show()
```

```python
# covering South America
h3 = h
for i in range(75, 150):
    for j in range(275, 325):
        h3[i, j] = -10000
plt.imshow(h3, interpolation = 'nearest')
plt.show()
```

```python
# exaggerating elevations
h4 = np.loadtxt('topography_180x360_grid.txt')
for i in range(0, 180):
    for j in range(0, 360):
        if h4[i, j] > 2000:
            h4[i, j] = 10000
        elif h4[i, j] < -2000:
            h4[i, j] = -10000
plt.imshow(h4, interpolation = 'nearest')
plt.show()
```
