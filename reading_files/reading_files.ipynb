```python
# plot everyone's data file as a thin black line
folder = 'class_data'
nFiles = 0;
names = []
for root, dirs, files in os.walk(folder):
    for filename in files:
        if (filename.find('.txt')>=0):
            xi = np.loadtxt(folder+'/'+filename)
            nFiles += 1;
            names.append(filename)
plt.figure(figsize = (10, 5))
for i in range(0, len(names)):
    plt.plot(np.loadtxt('class_data/' + names[i]), 'k')

# overplot solutions
plt.plot(solutions, 'r', linewidth = 2)
plt.show()
```

```python
# plot the deviation of every single data set 
# from the solution in one diagram
folder = 'class_data'
nFiles = 0;
names = []
for root, dirs, files in os.walk(folder):
    for filename in files:
        if (filename.find('.txt')>=0):
            xi = np.loadtxt(folder+'/'+filename)
            nFiles += 1;
            names.append(filename)
            
deviations = []
for i in range(0, len(names)):
    download = np.loadtxt('class_data/' + names[i])
    deviations.extend(download - solutions)
plt.figure(figsize = (10, 5))
plt.plot(deviations)
plt.show()
```

The most positive and the most negative deviations are concentrated a little past the 4000th mark. The student who collected the dataset shown there may have entered the data values in the wrong cell, since some values are off by more than 0.75.

```python
# variance between every data set and the solution
folder = 'class_data'
nFiles = 0;
names = []
var = []
for root, dirs, files in os.walk(folder):
    for filename in files:
        if (filename.find('.txt')>=0):
            xi = np.loadtxt(folder+'/'+filename)
            nFiles += 1;
            names.append(filename)
            
for i in range(0, len(names)):
    download = np.loadtxt('class_data/' + names[i])
    diffs = solutions - download
    var.append(np.var(diffs))

var
```

```python
# plot the variances
plt.plot(var)
plt.show()
```

```python
# determine the file name of the data set with the largest variance
names[np.argmax(var)]
```

```python
# determine the file name of the data set with the smallest variance
names[np.argmin(var)]
```

```python
# rank of my data set in an ordered list of all variances
names_with_var = [[names[i], var[i]] for i in range(0, len(var))]
ans = sorted(names_with_var, key = lambda x:x[1])
for i in range(0, len(ans)):
    if ans[i][0] == '5364615.txt':
        print(i)
```
